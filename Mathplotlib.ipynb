```python
import matplotlib.pyplot as plt
import numpy as np 
import pandas as pd
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
plt.figure(figsize=(5,3), dpi=100)

#plt.plot(x,y, label='2x', color='red', linewidth=2, marker=".", markersize = 10, linestyle = "--")
plt.plot(x,y,'b^--', label='2x')

x2 = np.arange(1,4.5,0.5)
plt.plot(x2[:3], x2[:3]**2, 'r', label='x^2')
plt.plot(x2[2:], x2[2:]**2, 'r--')

plt.title('Our first graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([1,2,3])
#plt.yticks([2,4,6,8])
plt.legend()

```

```python
labels = ['A', 'B', 'C']
values = [1,4,2]
bars = plt.bar(labels, values)
bars[0].set_hatch('/')
bars[1].set_hatch('o')
bars[2].set_hatch('*')
plt.show()
plt.figure(figsize=(6,4))
```
